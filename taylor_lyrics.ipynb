# Getting the lyrics for the album Lover 

```
lyrurl=[
        'https://www.lyrics.com/lyric/36467499/Taylor+Swift/I+Forgot+That+You+Existed',
        'https://www.lyrics.com/lyric/36467503/Shawn+Mendes/Lover' ,
        'https://www.lyrics.com/lyric/36497125/Taylor+Swift/Cruel+Summer',
        'https://www.lyrics.com/lyric/36467502/Taylor+Swift/The+Man',
        'https://www.lyrics.com/lyric/36467501/Taylor+Swift/The+Archer',
        'https://www.lyrics.com/lyric/36467500/Taylor+Swift/I+Think+He+Knows',
        'https://www.lyrics.com/lyric/36467504/Taylor+Swift/Miss+Americana+%26+the+Heartbreak+Prince',
        'https://www.lyrics.com/lyric/36497119/Taylor+Swift/Paper+Rings',
        'https://www.lyrics.com/lyric/36497118/Taylor+Swift/Cornelia+Street',
        'https://www.lyrics.com/lyric/36467490/Taylor+Swift/Death+by+a+Thousand+Cuts',
        'https://www.lyrics.com/lyric/36497116/Taylor+Swift/London+Boy',
        'https://www.lyrics.com/lyric/36467494/Dixie+Chicks/Soon+You%27ll+Get+Better',
        'https://www.lyrics.com/lyric/36497114/Taylor+Swift/False+God',
        'https://www.lyrics.com/lyric/36467487/Taylor+Swift/You+Need+to+Calm+Down',
        'https://www.lyrics.com/lyric/36467497/Taylor+Swift/Afterglow',
        'https://www.lyrics.com/lyric/36467496/Brendon+Urie/Me%21',
        'https://www.lyrics.com/lyric/36467493/Taylor+Swift/It%27s+Nice+to+Have+a+Friend',
        'https://www.lyrics.com/lyric/36467495/Taylor+Swift/Daylight',
        
      ]
```

```
#importing nltk which for preprocessing the lyrics
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize,  sent_tokenize
```

```
results = []

for link in lyrurl:
    try:
        song=requests.get(link)
        soup=BeautifulSoup(song.text,  'lxml')

        song_header=soup.find('h1' , {"class":"lyric-title"})
        title=song_header.text.strip()
        print(title)

        song_lyr=soup.find("pre" ,{"class":"lyric-body"})
        lyrics=song_lyr.text.strip()
        word = [word_tokenize(l) for  l in sent_tokenize(lyrics)]
        print(word)

    except TypeError:
        continue
    results.append((title,word))

with open('lover.csv' ,'w' )as csv_file:
    writer=csv.writer(csv_file)
    for  result in results:
        writer.writerow(result)
```
